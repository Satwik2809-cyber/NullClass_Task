```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")
```

```python
text = """
At Nullclass, we are committed to helping students gain real-world experience.
Our internship program focuses on AI/ML projects that prepare learners for industry-level challenges.
Through guided tasks and model implementation, interns sharpen their technical and research skills.
"""
tokens = tokenizer.tokenize(text)
print(f"Total Tokens ({len(tokens)}):", tokens)
```

```python
encoding = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
print("\nInput IDs:", encoding["input_ids"])
print("\nAttention Mask:", encoding["attention_mask"])
```

```python
with torch.no_grad():
    output = model(**encoding)
print("\nLast Hidden State Shape:", output.last_hidden_state.shape)
cls_vector = output.last_hidden_state[0][0]
print("\n[CLS] Embedding Vector (summary):", cls_vector)
```
